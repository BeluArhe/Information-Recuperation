### Paso 1: Calcular la matriz TF-IDF para el corpus `data/01_corpus_turismo_500.txt`

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. Cargar el corpus
with open('C:\\Users\\ASUS\\Documents\\7moSemestre\\RI\\Materia\\Information-Recuperation\\data\\01_corpus_turismo_500.txt', 'r', encoding='utf-8') as f:
    corpus_turismo = f.readlines()

# 2. Inicializar el vectorizador (puedes añadir stop_words='spanish')
vectorizer = TfidfVectorizer()

# 3. Calcular la matriz TF-IDF
tfidf_matrix_turismo = vectorizer.fit_transform(corpus_turismo)

# Opcional: Ver las primeras filas en un DataFrame para inspección
nombres_terminos = vectorizer.get_feature_names_out()
df_turismo = pd.DataFrame(tfidf_matrix_turismo.toarray(), columns=nombres_terminos)
print(df_turismo.head())
```

### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
import os 

# Usamos 'r' antes de la cadena para evitar problemas con las barras invertidas en Windows
path = r"C:\Users\ASUS\Documents\7moSemestre\RI\IB\ir26a\libros_gutenberg_es"
files = os.listdir(path)
len(files)

```

1. Cargar el contenido de los libros al Corpus

```python
corpus_gutenberg = []

for file_name in files:
    if file_name.endswith(".txt"):
        file_path = os.path.join(path, file_name)
        # Usamos errors='ignore' o 'replace' para evitar que el proceso se detenga
        with open(file_path, 'r', encoding='utf-8', errors='replace') as f:
            corpus_gutenberg.append(f.read())

print(f"Corpus creado con {len(corpus_gutenberg)} documentos.")
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

```python
import nltk
from nltk.corpus import stopwords

# Descargar las stop words 
nltk.download('stopwords')

# Crear la lista para español
stop_words_es = stopwords.words('spanish')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Configurar el vectorizador pasando la lista explícita
vectorizer_es = TfidfVectorizer(
    stop_words=stop_words_es, # Ahora pasamos la lista de NLTK
    max_df=0.85, 
    min_df=5
)

# 2. Calcular la matriz TF-IDF
tfidf_matrix_es = vectorizer_es.fit_transform(corpus_gutenberg)

# 3. Verificación de dimensiones
print(f"Documentos procesados: {tfidf_matrix_es.shape[0]}")
print(f"Vocabulario único (términos): {tfidf_matrix_es.shape[1]}")
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
from sklearn.metrics.pairwise import cosine_similarity

def buscar(query, vectorizer, tfidf_matrix, filenames, top_n=5):
    # Transformar la consulta al mismo espacio vectorial
    query_vec = vectorizer.transform([query])
    
    # Calcular similitud de coseno
    similitudes = cosine_similarity(query_vec, tfidf_matrix).flatten()
    
    # Obtener índices ordenados por relevancia
    indices_top = similitudes.argsort()[::-1][:top_n]
    
    print(f"🔎 Resultados para: '{query}'")
    for i in indices_top:
        if similitudes[i] > 0:
            print(f"Similitud: {similitudes[i]:.4f} | Archivo: {filenames[i]}")

# Prueba con una búsqueda temática
buscar("quijote sancho molinos", vectorizer_es, tfidf_matrix_es, files)
```
